1. In this assignment, you test whether the central limit theorem works. You generate 1000 variables with two normal distributions. You can determine the mean and standard deviation of these variables yourself. All you have to do is generate the first variable 50 times and averaged it each time. Generate the second variable 1000 times and averages this variable each time. Then plot the histogram of the averages of the two variables. Which of the variables has a mean distribution closer to the normal distribution? Do you think the Central Limit Theorem seems to have worked?

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statistics
from scipy.stats import norm
import seaborn as sns

mean1=20
mean2=40
sd1=10
sd2=5
sample_size=1000
normal_distribution1=np.random.normal(mean1,sd1,sample_size)
normal_distribution2=np.random.normal(mean2,sd2,sample_size)
```

```python
plt.figure(figsize=(8,5))
plt.hist(normal_distribution1,bins=20,alpha=0.5,label='Population 1')
plt.hist(normal_distribution2,bins=20,alpha=0.5,label='Population 2')
plt.axvline(mean1,color='darkblue',linestyle='solid',linewidth=3)
plt.axvline(mean2,color='darkorange',linestyle='solid',linewidth=3)
plt.title("Histogram \nPopulation 1 & Population 2", color="navy",fontsize=16)
plt.legend(loc='best');
```

```python
def sample_plot(pop,size,replication,color_sel,subplot_no):
    import numpy as np
    import seaborn as sns
    sample=[]
    sample_mean=[]
    sample_mean_shot=[]
    for number_shots in range(1,(replication+1)):
        shot=np.random.choice(pop,size,replace=True)
        sample.append(shot)
        sample_mean.append(np.mean(sample))
        sample_mean_shot.append(np.mean(shot))
        
    plt.subplot(subplot_no)
    plt.hlines(np.mean(pop),0,replication,colors='black')
    plt.plot(sample_mean,color=color_sel)
    plt.title('Sample Size={} and Replication={} \n Normal Distribution N({:2.0f},{:2.0f}) '.format(size,replication,np.mean(pop),np.std(pop)),fontsize=16)
    plt.xlabel('Replications',fontsize=14)
    plt.ylabel('Sample Mean',fontsize=14)
    
    plt.subplot(subplot_no+2)
    sns.distplot(sample_mean_shot, fit=norm, kde=False,bins=20, fit_kws=dict(linewidth=2.5,color=color_sel,label="Normal Fit"))
    plt.axvline(np.mean(pop),color='black',linestyle='solid',linewidth=3)   
    return plt
```

```python
plt.figure(figsize=(25,15))
sample_plot(pop=normal_distribution1,size=100,replication=50,color_sel="red",subplot_no=221)
sample_plot(pop=normal_distribution2,size=100,replication=1000,color_sel="red",subplot_no=222)
plt.show()
```

Central Tendecy Theorem states that if you have a population with mean $\mu$ and standard deviation $\sigma$ and take sufficiently large random samples from the population with replacement, then the distribution of the sample means will be approximately normally distributed.  

* We can conclude that the central tendency theorem seems to have worked as the number of replications increased. It is seen that the distribution of sample means related to second normal distribution looks much closer to normal distribution.
