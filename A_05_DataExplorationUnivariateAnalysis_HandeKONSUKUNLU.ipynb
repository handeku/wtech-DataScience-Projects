```python
import pandas as pd
import numpy as np
states = pd.read_csv("../../data/EDA/states_all.csv",parse_dates=['YEAR'])
```

Since most of the variables are right skewed and most of them have outliers, median imputation is more meaningful for this dataset.

```python
var_null=states.isnull().sum()
var_missing=var_null[var_null!=0].index
var_missing

states_median_imputation=states.copy()

for i in range(0,(len(var_missing))): 
    states_median_imputation[var_missing[i]].fillna(states_median_imputation[var_missing[i]].median(),inplace=True)

states_median_imputation.isnull().sum()*100/states_median_imputation.shape[0]
```

**(5)** You may have noticed that there are many missing values in the mathematics and reading notes. Complete the missing values with mean, median, and interpolation. Compare the impact of these techniques on the distribution of grades.

```python
math_reading_grades.isnull().sum()*100/math_reading_grades.shape[0]
```

```python
math_reading_grades_mean_imputation=math_reading_grades.copy()

for i in range(0,math_reading_grades.shape[1]): 
    math_reading_grades_mean_imputation.iloc[:,i].fillna(math_reading_grades_mean_imputation.iloc[:,i].mean(), inplace=True)

math_reading_grades_median_imputation=math_reading_grades.copy()

for i in range(0,math_reading_grades.shape[1]): 
    math_reading_grades_median_imputation.iloc[:,i].fillna(math_reading_grades_median_imputation.iloc[:,i].median(), inplace=True)
    
math_reading_grades_interpolate=math_reading_grades.copy()

for i in math_reading_grades_interpolate.columns: 
    math_reading_grades_interpolate[i] = math_reading_grades_interpolate[i].interpolate(method ='linear', limit_direction ='both')
    

```

```python

math_reading_grades_interpolate.isnull().sum()*100/math_reading_grades_interpolate.shape[0]

```

```python
compare_result=pd.DataFrame()
compare_result['Missing Rate']=math_reading_grades.isnull().sum()*100/math_reading_grades.shape[0]
compare_result['Before Imputation']=math_reading_grades.mean(axis=0)
compare_result['After Mean Imputation (Mean)']=math_reading_grades_mean_imputation.mean(axis=0)    
compare_result['After Median Imputation (Mean)']=math_reading_grades_median_imputation.mean(axis=0)   
compare_result['After Interpolation (Mean)']=math_reading_grades_interpolate.mean()
compare_result
```
