1. Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../../data/matplotlib_assignents/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
df.head()
df2=df[(df["location"].isin(["Spain","France","Germany","Italy"])) & (df['date'] > '2020-4-1')]
plt.figure(figsize=(20, 9))
sns.lineplot(x = 'date', y = 'total_deaths', data=df2, hue="location", lw=4, hue_order=["France","Germany","Italy","Spain"],color=["blue","orange","green","red"])
plt.title("Total Deaths", c="darkred", fontsize=20)
plt.xlabel('Dates',c="darkblue",fontsize=14)
plt.ylabel('Count of Deaths',c="darkblue",fontsize=14)
plt.legend()
plt.xticks(rotation = 45, fontsize = 9)
plt.show()
```

2. Plot a bar plot with seaborn for average death number that compares continents.

```python
plt.figure(figsize=(12, 8))
sns.barplot(x = 'continent', y = "total_deaths", order=["Asia","Europe","Africa","North America","South America","Ocenia"], data = df)
plt.xlabel("Continents", color="darkblue",fontsize=16)
plt.ylabel("Average Count of Deaths", color="darkblue",fontsize=16)
plt.title("Total Deaths", color="darkred",fontsize=24)
plt.show()
```

3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
covid_Turkey=pd.DataFrame(df[df['location']=="Turkey"][["new_deaths"]])
```

```python
plt.figure(figsize=(20, 16))

plt.subplot(2,2,1)
sns.distplot(covid_Turkey.new_deaths.dropna(), kde=False, bins=50, color = 'steelblue')
plt.title("New Deaths \n(Bins:50, KDE = False)",color="darkred",fontsize=20)
plt.xlabel("New Deaths", color="darkblue",fontsize=14)
plt.ylabel("Count", color="darkblue",fontsize=14)

plt.subplot(2,2,2)
sns.distplot(covid_Turkey.new_deaths.dropna(), kde = False, bins=20, color = 'steelblue')
plt.title("New Deaths \n(Bins:20, KDE = False)",color="darkred",fontsize=20)
plt.xlabel("New Deaths", color="darkblue",fontsize=14)
plt.ylabel("Count", color="darkblue",fontsize=14)

plt.subplot(2,2,3)
sns.distplot(covid_Turkey.new_deaths.dropna(), kde=True, bins=50, color = 'steelblue')
plt.title("New Deaths \n(Bins:50, KDE = True)",color="darkred",fontsize=20)
plt.xlabel("New Deaths", color="darkblue",fontsize=14)
plt.ylabel("Normalized Count", color="darkblue",fontsize=14)

plt.subplot(2,2,4)
sns.distplot(covid_Turkey.new_deaths.dropna(), kde = True, bins=20, color = 'steelblue')
plt.title("New Deaths \n(Bins:20, KDE = True)",color="darkred",fontsize=20)
plt.xlabel("New Deaths", color="darkblue",fontsize=14)
plt.ylabel("Normalized Count", color="darkblue",fontsize=14)

plt.show()
```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
covid_2C=pd.DataFrame(df[df['location'].isin(["Turkey","Italy"])][["location","new_deaths"]])
```

```python
plt.figure(figsize=(18, 7))

plt.subplot(1,3,1)
sns.boxplot(x="location", y="new_deaths", data=covid_2C)
plt.title("Boxplot",color="darkred",fontsize=16)
plt.xlabel("Country", color="darkblue",fontsize=12)
plt.ylabel("New Deaths", color="darkblue",fontsize=12)

plt.subplot(1,3,2)
sns.violinplot(x="location", y="new_deaths", data=covid_2C)
plt.title("Violin plot",color="darkred",fontsize=16)
plt.xlabel("Country", color="darkblue",fontsize=12)
plt.ylabel("New Deaths", color="darkblue",fontsize=12)

plt.subplot(1,3,3)
sns.swarmplot(x="location", y="new_deaths", data=covid_2C)
plt.title("Swarm plot",color="darkred",fontsize=16)
plt.xlabel("Country", color="darkblue",fontsize=12)
plt.ylabel("New Deaths", color="darkblue",fontsize=12)

plt.show()
```
