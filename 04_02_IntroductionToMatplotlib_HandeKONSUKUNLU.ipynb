1. Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3$\pi$ and 3$\pi$ with 0.1 intervals.

<font color=red>Let's import libraries:</font>


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-3*np.pi, 3*np.pi, 0.1)
y = np.cos(x)
```

```python
plt.plot(x, y, c= "#FF914D")
plt.title("Cosine Function Plot")
plt.show()
```

2. This time, create the same plot in red color with object methods. 

```python
fig=plt.figure()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, color = 'red')
plt.title("Cosine Function Plot")
plt.show()
```
