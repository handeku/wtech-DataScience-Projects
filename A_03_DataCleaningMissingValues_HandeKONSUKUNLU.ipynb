**(2)** You may have noticed that our data has a year column. For now, forget about the year data and assume that each observation was made in the same year. Consider how you can fill in missing values for each variable. Filling in the missing values with a value is meaningful for which variables and for which ones?

```python
var_null=states.isnull().sum()
var_missing=var_null[var_null!=0].index
var_missing
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
j=1
plt.figure(figsize=(40,32))
for i in range(0,len(var_missing)): 
    plt.subplot(8,3,j)
    plt.hist(states[var_missing[i]],color="green")
    plt.title("{}".format(var_missing[i]), color="darkblue",fontsize=16)
    j=j+1
  
```

```python
j=1
plt.figure(figsize=(40,45))
for i in range(0,len(var_missing)): 
    plt.subplot(8,3,j)
    sns.boxplot(states[var_missing[i]],color="green")
    plt.title("{}".format(var_missing[i]), color="darkblue",fontsize=16)
    j=j+1
  
```

INTERPRETATION: As it was seen from the histogram and boxplots, most of the variables are right skewed. In addition, they have outliers. Taking into consideration the mising value rates, filling in the missing values with a value is meaningful for the variables 'ENROLL',  'TOTAL_REVENUE', 'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE', 'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 
'SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE', 'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G', 'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G', 'GRADES_9_12_G', 'GRADES_ALL_G'.

However the variables 'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE', 'AVG_READING_4_SCORE', 'AVG_READING_8_SCORE' have very high missing rate. Therefore it is not meaningful for these variables.

Now, I will fill in the missing values with 'median' instead of mean since the distribution of the variables are right-skewed and these variables have outliers.

```python
states_mean_imputation=states.copy()

for i in range(0,(len(var_missing)-4)): 
    states_mean_imputation[var_missing[i]].fillna(states_mean_imputation[var_missing[i]].mean(), inplace=True)

states_median_imputation=states.copy()

for i in range(0,(len(var_missing)-4)): 
    states_median_imputation[var_missing[i]].fillna(states_median_imputation[var_missing[i]].median(),inplace=True)
```

Let's check the missing rates for the variables which were imputed:

```python
missing_result1=pd.DataFrame()
missing_result1['Before_Imputation']= states.isnull().sum()*100/states.shape[0]
missing_result1['Mean_Imputation']= states_mean_imputation.isnull().sum()*100/states_mean_imputation.shape[0]
missing_result1['Median_Imputation']= states_median_imputation.isnull().sum()*100/states_median_imputation.shape[0]
missing_result1
```

```python
imputation_result1=pd.DataFrame()
imputation_result1['Before Imputation (Mean)']= states.mean(axis=0).iloc[:-4]
imputation_result1['After Mean Imputation (Mean)']= states_mean_imputation.mean(axis=0)[:-4]    
imputation_result1['After Median Imputation (Mean)']= states_median_imputation.mean(axis=0)[:-4]   
imputation_result1
```

INTERPRETATION: As we can see, the median imputation is more meaningful than mean imputation.

**(3)** Now it's time to consider the factor of time! Review your answer in question 2 and fill in the missing data based on the values observed during that year. For example, if you want to fill a value with an average value, calculate the average of that year.

```python
states['YEARS']=states['YEAR'].dt.year
states.head()
```

```python
variables=states.columns[3:-1]
                 
states_mean_year_imputation=states.copy()
states_median_year_imputation=states.copy()

for j in variables:
    states_mean_year_imputation[j] = states_mean_year_imputation[j].fillna(states_mean_year_imputation.groupby("YEARS")[j].transform('mean'))
    states_median_year_imputation[j]=states_median_year_imputation[j].fillna(states_median_year_imputation.groupby("YEARS")[j].transform('median'))
  
```

```python
variables=states.columns[3:-1]
variables
```

```python
imputation_result2=pd.DataFrame()
imputation_result2['Before Imputation (Mean) ']= states.median(axis=0).iloc[:-1]
imputation_result2['After Mean Imputation (Mean) by Year']= states_mean_year_imputation.mean(axis=0).iloc[:-1]  
imputation_result2['After Median Imputation (Mean) by  Year']= states_median_year_imputation.mean(axis=0).iloc[:-1]   
imputation_result2
```

**(4)** This time, fill in the missing values by interpolating.

```python
variables=states.columns[3:-1]
states_interpolate=states.copy()

for j in variables:
    states_interpolate[j] = states[j].interpolate(method ='linear', limit_direction ='forward')
```

```python
imputation_result3=pd.DataFrame()
imputation_result3['Before Imputation (Mean)']= states.mean().iloc[:-1]
imputation_result3['After Interpolation (Mean)']= states_interpolate.mean().iloc[:-1]    
imputation_result3
```
