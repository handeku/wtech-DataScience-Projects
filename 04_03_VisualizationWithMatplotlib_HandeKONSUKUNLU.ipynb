In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

covid = pd.read_csv("../../data/matplotlib_assignents/owid-covid-data.csv",parse_dates=['date'])
covid.head()
```

```python
covid_Spain=covid[covid.location=='Spain'].sort_values(by=['date'])
covid_Spain_last100=covid_Spain[-100:]
covid_France=covid[covid.location=='France'].sort_values(by=['date'])
covid_France_last100=covid_France[-100:]
covid_Germany=covid[covid.location=='Germany'].sort_values(by=['date'])
covid_Germany_last100=covid_Germany[-100:]
covid_Italy=covid[covid.location=='Italy'].sort_values(by=['date'])
covid_Italy_last100=covid_Italy[-100:]
```

```python
covid_4C=pd.concat([covid_Spain_last100,covid_France_last100,covid_Germany_last100,covid_Italy_last100])
```

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

```python
plt.figure(figsize=(25,5))
plt.subplot(1,2,1)
plt.plot(covid_4C[covid_4C['location']=="Spain"]['date'],covid_4C[covid_4C['location']=="Spain"]['total_cases'],c="orange")
plt.plot(covid_4C[covid_4C['location']=="France"]['date'],covid_4C[covid_4C['location']=="France"]['total_cases'],c="green")
plt.plot(covid_4C[covid_4C['location']=="Germany"]['date'],covid_4C[covid_4C['location']=="Germany"]['total_cases'],c="red")
plt.plot(covid_4C[covid_4C['location']=="Italy"]['date'],covid_4C[covid_4C['location']=="Italy"]['total_cases'],c="blue")
plt.legend(["SPAIN","FRANCE","GERMANY","ITALY"])
plt.subplot(1,2,2)
plt.plot(covid_4C[covid_4C['location']=="Spain"]['date'],covid_4C[covid_4C['location']=="Spain"]['total_deaths'],c="orange")
plt.plot(covid_4C[covid_4C['location']=="France"]['date'],covid_4C[covid_4C['location']=="France"]['total_deaths'],c="green")
plt.plot(covid_4C[covid_4C['location']=="Germany"]['date'],covid_4C[covid_4C['location']=="Germany"]['total_deaths'],c="red")
plt.plot(covid_4C[covid_4C['location']=="Italy"]['date'],covid_4C[covid_4C['location']=="Italy"]['total_deaths'],c="blue")
plt.legend(["SPAIN","FRANCE","GERMANY","ITALY"]);
```
