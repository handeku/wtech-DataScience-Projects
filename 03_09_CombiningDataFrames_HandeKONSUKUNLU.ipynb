````python
features=pd.read_csv("C:/Users/HANDEKONSUKUNLU/Documents/Data Science Course-wtech/data/pandas_assignments/features.txt")
features
````

````python
target=pd.read_csv("C:/Users/HANDEKONSUKUNLU/Documents/Data Science Course-wtech/data/pandas_assignments/target.txt")
target
````

<font color=red> Merge these two datasets by using **inner join**.</font>

````python
df_inner=pd.merge(features, target, on='id')
df_inner
````

````python
print("Merged dataset by using inner join has {} rows and {} columns.".format(df_inner.shape[0],df_inner.shape[1]))
````

<font color=red> Merge these two datasets by using **outer join**.</font>

````python
df_outer=pd.merge(features, target, on='id', how='outer')
df_outer
````

````python
print("Merged dataset by using outer join has {} rows and {} columns.".format(df_outer.shape[0],df_outer.shape[1]))
````

<font color=red> Merge these two datasets by using **left join**.</font>

````python
df_left=pd.merge(features, target, on='id', how='left')
df_left
````

````python
print("Merged dataset by using left join has {} rows and {} columns.".format(df_left.shape[0],df_left.shape[1]))
````

<font color=red> Merge these two datasets by using **right join**.</font>

````python
df_right=pd.merge(features, target, on='id', how='right')
df_right
````

````python
print("Merged dataset by using right join has {} rows and {} columns.".format(df_right.shape[0],df_right.shape[1]))
````
